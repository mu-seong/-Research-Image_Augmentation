# Dissim

```python
img_folders_path = 'C:/Users/user/0. working/1. Research/1. oriental/data/9. styleimg_label_matching/img/Dissim/'
img_folders_path
```

```python
img_folders = os.listdir(img_folders_path)
img_folders
```

```python
new_img_path = 'C:/Users/user/0. working/1. Research/1. oriental/Dataset/Dissim/train/images/'
new_img_path
```

```python
label_folders_path = 'C:/Users/user/0. working/1. Research/1. oriental/data/9. styleimg_label_matching/label/Dissim/'
label_folders_path
```

```python
label_folders = os.listdir(label_folders_path)
label_folders
```

```python
new_label_path = 'C:/Users/user/0. working/1. Research/1. oriental/Dataset/Dissim/train/labels/'
new_label_path
```

# img & label

```python
i = 1

for img_folder, label_folder in zip(img_folders, label_folders):
    img_objects_path = img_folders_path + img_folder
    img_objects = os.listdir(img_objects_path)
    
    label_objects_path = label_folders_path + label_folder
    label_objects = os.listdir(label_objects_path)
    
    for img_obj, label_obj in zip(img_objects, label_objects):
        img_files_path = img_objects_path + '/' + img_obj
        img_files = os.listdir(img_files_path)
        
        label_files_path = label_objects_path + '/' + label_obj
        label_files = os.listdir(label_files_path)
        
        for idx, (img_file, label_file) in enumerate(zip(img_files, label_files)):
            if img_file.split('.')[0][6:] == label_file.split('.')[0]:
                print(img_file, label_file)
                img_src = os.path.join(img_files_path, img_file)
                img_dst = img_obj + '_style_' + str(i) + '_img_' + str(idx) + '.jpg'
                img_dst = os.path.join(new_img_path, img_dst)
                os.rename(img_src, img_dst)

                label_src = os.path.join(label_files_path, label_file)
                label_dst = label_obj + '_style_' + str(i) + '_img_' + str(idx) + '.txt'
                label_dst = os.path.join(new_label_path, label_dst)
                os.rename(label_src, label_dst)
            else:
                print('='*60)
            
    i += 1
```
